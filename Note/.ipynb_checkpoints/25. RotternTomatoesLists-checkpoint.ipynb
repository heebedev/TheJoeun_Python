```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd
```

```python
url = "https://www.rottentomatoes.com/top/bestofrt/?year=2019"
page = urlopen(url)
soup = BeautifulSoup(page,'html.parser')
```

```python
soup.select("div td a.articleLink")[0].string.strip()
```

```python
#예매율 영화제목 추출하기

li_lists = soup.select("div td a.articleLink")
count = 0
titleLists = [] #영화제목 Lists
countLists = [] #예매율 순위 Lists
yearLists = []

for li in li_lists :
    count += 1
    print(count, li.string.strip()[0:-7])
    countLists.append(count)
    year = li.string.strip()[-5:-1]
    yearLists.append(year)
    titleLists.append(li.string.strip()[0:-7])
```

```python
#countLists 만으로 Datafram 생성
df = pd.DataFrame(data = countLists, columns=['Index'])
```

```python
# 기존 dataframe에 컬럼과 List 추가
df['Titles'] = titleLists
df['Year'] = yearLists
df
```

```python
# csv wjwkd
df.to_csv("../Data/RottenList_2019.csv", index=False)
```

```python
ratings = soup.select("div td span.tMeterScore")
titles = soup.select("div td a.articleLink")
reviews = soup.select("div td.right")

count = 0
countLists = [] #예매율 순위 Lists
ratingLists = [] # 영화로튼지수 Lists
titleLists = [] #영화제목 Lists
reviewLists = [] #영화 리뷰 Lists

for title in titles :
    count += 1
    ra = ratings[titles.index(title)].string
    re = reviews[titles.index(title)].string
    print(count, ra, title.string.strip()[0:-7], re)
    countLists.append(count)
    ratingLists.append(ra)
    titleLists.append(title.string.strip()[0:-7])
    reviewLists.append(re)
```

```python
#countLists 만으로 Datafram 생성
df = pd.DataFrame(data = countLists, columns=['Index'])
```

```python
# 기존 dataframe에 컬럼과 List 추가
df['Titles'] = titleLists
df['Rating'] = ratingLists
df['Review'] = reviewLists
df
```

```python
# csv 저장
df.to_csv("../Data/RottenList_2019_DV.csv", index=False)
```
