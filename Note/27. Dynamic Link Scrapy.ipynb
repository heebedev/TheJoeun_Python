### Test completed

```python
urlSelects = soup.select("#top_movies_main > div > table > tr > td > a")
urlSelects
```

```python
genres = []
directors = []
count = 0

for i in urlSelects :
    count += 1
    urlSel = i['href']
    wkUrl = urlBase + urlSel
    pageSel = urlopen(wkUrl)
    soulSel = BeautifulSoup(pageSel, 'html.parser')
    try:
        directors.append(soulSel.select("li > div.meta-value a")[1].string)
        genres.append(soulSel.select("li > div.meta-value")[1].string.strip().replace(' ',"").replace('\n',''))
        print(count,":",soulSel.select("li > div.meta-value a")[1].string)
        print(count,":",soulSel.select("li > div.meta-value")[1].string.strip().replace(' ',"").replace('\n',''))
    except IndexError as e :
        directors.append("")
        genres.append(soulSel.select("li > div.meta-value")[1].string.strip().replace(' ',"").replace('\n',''))
        print(count,":",soulSel.select("li > div.meta-value")[1].string.strip().replace(' ',"").replace('\n',''))
        
print(genres)
print(directors)
```

```python
titles = []
count = 0

for i in urlSelects :
    count += 1
    urlSel = i['href']
    wkUrl = urlBase + urlSel
    pageSel = urlopen(wkUrl)
    soulSel = BeautifulSoup(pageSel, 'html.parser')
    print(count, ":", soulSel.select("h1.mop-ratings-wrap__title--top")[0].string)
    try:
        titles.append(soulSel.select("h1.mop-ratings-wrap__title--top")[0].string)
    except IndexError as e :
        directors.append("")
        
print(titles)
```

```python
url = "http://ncov.mohw.go.kr/"

#main_maplayout > button:nth-child(9)

pageSelect = urlopen(url)
soulSelect = BeautifulSoup(pageSelect, 'html.parser')
print(soulSelect)
```

```python
names = soulSelect.select("#main_maplayout button span.name")
nums = soulSelect.select("#main_maplayout button span.num")
befores = soulSelect.select("#main_maplayout button span.before")

name = []
num = []
before = []

for i in names :
    name.append(i.string)
    num.append(nums[names.index(i)].string)
    before.append(befores[names.index(i)].string)
                    
print(name)
print(num)
print(before)
```

```python
index1 = list(range(1, len(names)+1))
print(index1)
```

```python
df1 = pd.DataFrame(data = index1, columns=['Index'])
df1
```

```python
df1['Name'] = name
df1['Num'] = num
df1['Before'] = before
df1
```
