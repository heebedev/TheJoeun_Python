### 컬럼 추가하기 & 수정하기

```python
train['Category'] = 'Titanic'
```

```python
train['Id'] = range(891)
train.head(3)
```

```python
# FamilySize 컬럼에 Family Size 구하기

# SibSp + Parch + 1(자기자신)

train['FamilySize'] = train['SibSp'] + train['Parch'] + 1
```

```python
train.head(3)
```

```python
train['FamilySize'].max()
```

```python
# Embarked가 C인 것과 아닌것에 대한 컬럼 추가하기
# Chersbourg(FR), Southapton(UK), QueensTown(UK)

train['Nationality_FR'] = train['Embarked'] == "C"
train['Nationality_UK'] = train['Embarked'] != "C"
train.head(3)
```

```python
# Fare에 대한 기본통계량 보기
train['Fare'].describe()
```

```python
# 분류 : 2분위수와 3분위수의 근사 정수치를 선택

train['LowFare'] = train['Fare'] <15
train['MediumFare'] = (train['Fare'] >= 15) & (train['Fare'] < 30)
train['HighFare'] = train['Fare'] >= 30

train[['Fare', 'LowFare', 'MediumFare', 'HighFare']].head(10)
```

```python
# Fare에 대한 하나의 컬럼으로 분류하기

train.loc[train['Fare'] < 15, 'FareType'] = 'Low'
train.loc[(train['Fare'] >= 15) & (train['Fare'] < 30), "FareType"] = 'Medium'
train.loc[train['Fare'] > 30, "FareType"] = "High"

train[["Fare", "FareType"]].head(10)
```

```python
# Age의 NaN을 Age의 평균으로 대체하기

train.loc[train['Age'].isnull(), "Age"] = round(train['Age'].mean(),1)
```
